```python

import warnings
warnings.filterwarnings('ignore')

import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
import seaborn as sns
import os

```

Let's read and store all the csv files.

```python
batting_Data = pd.read_csv('batting.csv')
bowling_Data = pd.read_csv('bowling.csv')
fielding_Data = pd.read_csv('fielding.csv')
results_Data = pd.read_csv('results.csv')
cpl_Data = pd.read_csv('cpl.csv')
```

**Regarding Team-Wise Results Data**

```python
results_Data
```

We can observe that India and Newzealand are the only two teams which has played more than 20 matches,in which Newzealand marchs towards 30.Other teams have also played a significant amount of matches. Since it was a 50-Over world cup happeded last year,i.e, 2023 in India,the amount of T20s played in the span of two years is actually less.

Let us plot some graphs with this data.

```python

X = results_Data["Team"]
fig=plt.figure(figsize=(15,8))
ax=fig.add_axes([0,0,1,1])
ax.bar(X , results_Data["Won"], color = 'b', width = 0.25)
ax.bar(X , results_Data["Lost"], color = 'g', width = 0.25)
ax.bar(X , results_Data["Tied"], color = 'r', width = 0.25)

ax.set_xticks([0.25,1.25,2.25,3.25,4.25,5.25,6.25,7.25,8.25,9.25])
ax.set_xticklabels(['India','Newzealand','Australia','West Indies','Bangladesh','Afghanistan','Sri Lanka','England','Pakistan','South Africa'])
ax.legend(labels=['WON','LOST','TIED'])
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
ax=fig.add_axes([0,0,1,1])
Team=results_Data["Team"]
Ratio=results_Data["W/L"]

ax.bar(Team,Ratio,color='g')
plt.show()
```

From this,we get to know India and australia have high Win-loss Ratio among all other teams

```python
fig=plt.figure(figsize=(15,8)) 
ax=fig.add_axes([0,0,1,1])
ax.axis('equal')
Team=results_Data["Team"]
rpo=results_Data["RPO"]
ax.pie(rpo, labels=Team,autopct='%1.2f%%')
plt.show()
```

From the above pie chart,Teams like South Africa,Australia and West Indies have Run rate per over around 11.50 which tells that they play more aggresive cricket than other teams.
